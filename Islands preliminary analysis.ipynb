# Aeroe

```r
## Read .csv dsata
wr <- read.table("Aeroe_timeseries_data.csv", header=TRUE, sep= "," ,as.is=TRUE)
head(wr,n=10)
tail(wr,n=5)
```

```r
#Tatol supply calculation
con_tot = vector() #Defining empty vector for holding total sumply vs. day values
for (row in 1:nrow(wr)){ #Iterating through rows of the array
    con_tot <- append(con_tot, wr[row,"solar_supply_kW"] + wr[row,"wind_supply_kW"])  #Appending array
    }
wr$total_supply_kW <- con_tot #Adding column to the dataframe

#Ploting
plot(wr$hr,wr$consumption_kW,type ="l", ylim=c(0,max(wr$consumption_kW)),xlim=c(0,1000),sub="Aero P consumption/production",xlab="hours",ylab="power",lwd=3)
lines(wr$hr,wr$solar_supply_kW,type="l",col="coral4",lwd=3)
lines(wr$hr,wr$wind_supply_kW,type="l",col="blue",lwd=3)
lines(wr$hr,wr$total_supply_kW,type="l",col="tomato2",lwd=3)

#Finding difference between supply and demand per day
options(repr.plot.width=28, repr.plot.height=10)
```

**Analysis**

```r
# Energy shortage
n <- 0 #Counter for amount of hors with shortage
for (row in 1:nrow(wr)){ #Iterating through rows of the array
    if (wr[row,"total_supply_kW"] < wr[row,"consumption_kW"]){
        n <- n +1
    }   
    }

# Share of renewable energy
cat("Total",sum(wr$total_supply_kW)/sum(wr$consumption_kW)*100," %\n")
cat("Solar",sum(wr$solar_supply_kW)/sum(wr$consumption_kW)*100," %\n")
cat("Wind",sum(wr$wind_supply_kW)/sum(wr$consumption_kW)*100," %\n")

cat("Shortage",n/max(cyprus$hr)*100," %\n")
```
